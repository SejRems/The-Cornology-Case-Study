```python
#Importing package to apply the linear programing. 
import pulp as p
import pandas as pd
import numpy as np


# Create variables for the demands 
#Xfg = #of units of flavor "f" for market "m"
# f = 1 for original flavor and 2 for blast flavor.
# g = 1 for Northeast, 2 for YRD, 3 for Southeast, 4 for North Center, 5 for South Central, and 6 for Western China.

x11_original = 0
x21_blast = 70
x12_original = 180 
x22_blast =  112
x13_original = 127
x23_blast = 37
x14_original = 90  
x24_blast = 70
x15_original = 118 
x25_blast = 74 
x16_original = 90
x26_blast = 0

# m1 is the minimum of units per run for original flavor (90). 
# m2 is the minimum of units per run for blast flavor (70). 
# We will use these variables to get the number of runs per market. 
#As a team, we identified that the more efficient way to maximize 
#our production is attending markets that are multiples of m1 and/or m2

m1=90
m2=70

##Making our variables divisible by the minimum units per flavor.
x11_new_original = x11_original/m1
x21_new_blast = x21_blast/m2
x12_new_original = x12_original/m1 
x22_new_blast =  x22_blast/m2
x13_new_original = x13_original/m1
x23_new_blast = x23_blast/m2
x14_new_original = x14_original/m1  
x24_new_blast = x24_blast/m2
x15_new_original = x15_original/m1 
x25_new_blast = x25_blast/m2
x16_new_original = x16_original/m1
x26_new_blast = x26_blast/m2

##Here we used the conditional to identified which market is divisible by the minimum units per flavor.
##Put 0 to the market which does not satisfy the condition. 

if x11_new_original.is_integer():
    x11_new_original == x11_new_original
else:
    x11_new_original= 0
    
if x21_new_blast.is_integer():
    x21_new_blast == x21_new_blast
else:
    x21_new_blast= 0
    
if x12_new_original.is_integer():
    x12_new_original == x12_new_original
else:
    x12_new_original= 0

if x22_new_blast.is_integer():
    x22_new_blast == x22_new_blast
else:
    x22_new_blast= 0
    
if x13_new_original.is_integer():
    x13_new_original == x13_new_original
else:
    x13_new_original= 0
    
if x23_new_blast.is_integer():
    x23_new_blast == x23_new_blast
else:
    x23_new_blast= 0

if x14_new_original.is_integer():
    x14_new_original = x14_new_original
else:
    x14_new_original= 0    

if x24_new_blast.is_integer():
    x24_new_blast = x24_new_blast
else:
    x24_new_blast= 0    

if x15_new_original.is_integer():
    x15_new_original == x15_new_original
else:
    x15_new_original= 0  

if x25_new_blast.is_integer():
    x25_new_blast == x25_new_blast
else:
    x25_new_blast= 0  

if x16_new_original.is_integer():
    x16_new_original == x16_new_original
else:
    x16_new_original= 0     

if x26_new_blast.is_integer():
    x26_new_blast = x26_new_blast
else:
    x26_new_blast= 0    
    
#We made a list just to see the big picture of the result. 

new_demand = [x11_new_original, x21_new_blast, x12_new_original,x22_new_blast ,x13_new_original ,x23_new_blast ,x14_new_original ,
              x24_new_blast ,x15_new_original, x25_new_blast,x16_new_original ,x26_new_blast]


##Difine variables:
##xfg = f for flavor (1 and 2) and m for location (1,2,3,4,5,and 6)

Max_proft = p.LpProblem('Maximize Profit', p.LpMaximize) 

#Difining variables for orginal flavor
x11 = p.LpVariable("x11", lowBound = 0)   # Create a variable >= 0
x12 = p.LpVariable("x12", lowBound = 0)   # Create a variable >= 0
x13 = p.LpVariable("x13", lowBound = 0)   # Create a variable >= 0
x14 = p.LpVariable("x14", lowBound = 0)   # Create a variable >= 0
x15 = p.LpVariable("x15", lowBound = 0)   # Create a variable >= 0
x16 = p.LpVariable("x16", lowBound = 0)   # Create a variable >= 0

#Difining variables for blast flavor
x21 = p.LpVariable("x21", lowBound = 0)   # Create a variable >= 0
x22 = p.LpVariable("x22", lowBound = 0)   # Create a variable >= 0
x23 = p.LpVariable("x23", lowBound = 0)   # Create a variable >= 0
x24 = p.LpVariable("x24", lowBound = 0)   # Create a variable >= 0
x25 = p.LpVariable("x25", lowBound = 0)   # Create a variable >= 0
x26 = p.LpVariable("x26", lowBound = 0)   # Create a variable >= 0

#Profit per flavor. 
p1 = 12500
p2= 7500

#Investment or cost of the factory in Yala = 10,303,601

inv = 10303601

# Objective Function
Max_proft += p1*m1*(x11 + x12 + x13 + x14 + x15 + x16) + p2*m2*(x21 + x22 + x23 + x24 + x25 + x26)- inv

# Constraints:
Max_proft += x11== x11_new_original
Max_proft += x12== x12_new_original
Max_proft += x13== x13_new_original
Max_proft += x14== x14_new_original
Max_proft += x15== x15_new_original
Max_proft += x16== 1                 #We know that the entry port of Yala is Western China. We must to attend the demand in this market.
Max_proft += x21== x21_new_blast
Max_proft += x22== x22_new_blast
Max_proft += x23== x23_new_blast
Max_proft += x24== x24_new_blast
Max_proft += x25== x25_new_blast
Max_proft += x26== x26_new_blast

Max_proft += x11 + x12+ x13+ x14+ x15+ x16 + x21 + x22+ x23+ x24+ x25+ x26 == 5  # the sum of the runs must be 5 to maximize our profit.

# Display the problem
print(Max_proft)

status = Max_proft.solve()   # Solver

  
# Printing the final solution
print(p.value(x11), p.value(x12),p.value(x13),
      p.value(x14),p.value(x15),p.value(x16),
      p.value(x21),p.value(x22),p.value(x23),
      p.value(x24),p.value(x25),p.value(x26))

print(p.value(Max_proft.objective))

###The best alternative based on our constraint is: Northeast (Blast) + YRD (Original) + North Central (Original) + Western China (Original)
```
